```python
import pickle
import numpy as np
import pandas as pd
from plotnine import *
```

```python
with open('wiki_wc.pickle', 'rb') as f:
    wc = pickle.load(f)
```

```python
x = np.asarray([x for x in wc.values()])
print(len(x))
#x = np.sort(x)
#x = x[:600]
df = pd.DataFrame({
    'wc': x
})
df['log_wc'] = df['wc'].apply(np.log)
```

```python
(
    ggplot(df, aes('log_wc'))
    + geom_histogram()
    + ggtitle('Distribution of log of word counts')
#    + theme_xkcd()
)
```
